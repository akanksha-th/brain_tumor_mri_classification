```python
import matplotlib.pyplot as plt, os, PIL, random
```

```python
dir_name = os.path.join('..', '.data', 'train')
tumor_classes = ["glioma", "meningioma", "no_tumor", "pituitary"]

plt.figure(figsize=(12, 4))
for i, tumor in enumerate(tumor_classes):
    folder = os.path.join(dir_name, tumor)
    imgs = [f for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp'))]
    if not imgs:
        continue
    img_file = random.choice(imgs)
    img = PIL.Image.open(os.path.join(folder, img_file))
    plt.subplot(1, len(tumor_classes), i + 1)
    plt.imshow(img)
    plt.title(tumor, fontdict={'weight': 'bold', 'fontsize': 14})
    plt.axis('off')
plt.tight_layout()
plt.show()
```
